```python
file_path = '/content/drive/MyDrive/Pandas_2025/data/transacoes.csv'
df = pd.read_csv(file_path)#.convert_dtypes()
df.head()
```

```python
# Schema
df.dtypes
```

```python
df.info(memory_usage='deep')
```

```python
# converter a coluna dtCriacao para datetime
df['dtCriacao'] = pd.to_datetime(df['dtCriacao'])
```

```python
# Pegar transações com mais de 50 pontos
filtro = (df['qtdePontos'] >= 50 )
df[filtro]
```

```python
# Mais de 50 e menor ou igual a 100
filtro = (df['qtdePontos'] >= 50 ) & (df['qtdePontos'] < 100)
df[filtro]
```

```python
# Pegar quantidade ponto igual a 1 OU igual a 100
df[(df['qtdePontos'] == 1) | (df['qtdePontos'] == 100)]
```

```python
# Entre maiores que 0 e menor que 50 do ano de 2025
df[(df['dtCriacao'].dt.year == 2025) & (df['qtdePontos'] > 0) & (df['qtdePontos'] <= 50)]
```
