```python
import pandas as pd
import numpy as np
```

### Operações com escalar

```python
file = '../data/clientes.csv'
clientes = pd.read_csv(file)
clientes.info()
```

```python
clientes.sample(10)
```

```python
# Somando 100 pontos para cada usuário fazendo operação escalar

clientes['qtdePontos'] + 100 # <- isso não é uma view e essa operação persiste a mudança nos dados. 
```

### Criando Novas Colunas - Create new Columns

```python
clientes['pontos_100'] = clientes['qtdePontos'] + 100
clientes
```

```python
clientes['flEmail'].value_counts()
```

```python
clientes['flTwitch'].value_counts()
```

```python
clientes['flYouTube'].value_counts()
```

```python
clientes['emailTwitch'] = clientes['flEmail'] + clientes['flTwitch']
clientes
```

```python
clientes['emailTwitch'].value_counts()
```

```python
clientes['qtSocial'] = clientes['flEmail']+clientes['flTwitch']+clientes['flYouTube']+clientes['flBlueSky']+clientes['flInstagram']
clientes['qtSocial'].value_counts()
```

```python
clientes['unlestOne'] = clientes['flEmail']*clientes['flTwitch']*clientes['flYouTube']*clientes['flBlueSky']*clientes['flInstagram']
clientes['unlestOne'].value_counts()
```

### Operações entre colunas


```python
clientes['qtdePontos'].describe()
```

```python
# Aplicando logaritmico com numpy

clientes['logPontos'] = np.log(clientes['qtdePontos']+1)

clientes['logPontos'].describe()
```

```python
import matplotlib.pyplot as plt

plt.hist(clientes['logPontos'])
plt.grid(True)
plt.show()
```

# Ordenação


```python
# Descobrindo quem fez mais pontos
maior_pontuacao = clientes['qtdePontos'].max()
clientes[clientes['qtdePontos'] == maior_pontuacao]
```

```python
# 5 maiores pontuações

clientes[clientes['qtdePontos'].isin(clientes['qtdePontos'].nlargest(5).tolist())]
```

```python
# Ordenando pela quantidade de pontos e filtrando as 5 maiores pontuacoes

top_5 = clientes.sort_values(by='qtdePontos', ascending=False).head(5)
top_5 
```
