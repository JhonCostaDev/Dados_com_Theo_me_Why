# Variana

Na rea da estatstica, a varincia  uma medida de disperso que indica o quo espalhados esto os valores de um conjunto de dados em relao  sua mdia (ou valor esperado). Em outras palavras, ela quantifica a variabilidade dos dados.

## Conceito Intuitivo:

Imagine que voc tem dois conjuntos de notas de alunos. Ambos os conjuntos podem ter a mesma mdia, mas em um conjunto as notas podem estar muito prximas da mdia (por exemplo, a maioria dos alunos tirou notas entre 7 e 8), enquanto no outro conjunto as notas podem estar mais dispersas (alguns alunos tiraram notas muito baixas e outros muito altas). A varincia ajuda a distinguir essas situaes, indicando que o segundo conjunto tem uma varincia maior.

## Como a Varincia  Calculada:

O clculo da varincia envolve os seguintes passos:

1. Calcular a mdia () do conjunto de dados.
2. Para cada valor (x) no conjunto de dados, calcular a diferena entre esse valor e a mdia (x - ). Este valor  chamado de desvio.
3. Elevar ao quadrado cada um desses desvios ((x - )). Isso  feito para garantir que as diferenas negativas e positivas contribuam para a medida de disperso e para dar mais peso aos desvios maiores.
4. Calcular a mdia desses desvios quadrados.

<hr>

## Frmulas:

Existem duas frmulas principais para calcular a varincia, dependendo se voc est trabalhando com uma populao inteira ou com uma amostra:

### Varincia Populacional ():

```bash
 =  (x - ) / N
```
Onde:

*  (sigma ao quadrado) representa a varincia populacional.
*  (sigma maisculo) indica a soma.
* x representa cada valor individual na populao.
*  (mi) representa a mdia da populao.
* N representa o nmero total de elementos na populao.


<hr>

### Varincia Amostral (s):

```bash
s =  (x - x) / (n - 1)

```
Onde:

* s representa a varincia amostral.
*  indica a soma.
* x representa cada valor individual na amostra.
* x (x-barra) representa a mdia da amostra.
* n representa o nmero total de elementos na amostra.
O denominador  (n - 1) em vez de n para fornecer uma estimativa no viesada da varincia populacional a partir da amostra (correo de Bessel).


## Interpretao da Varincia:

### Varincia baixa:
Indica que os dados esto agrupados prximos  mdia, ou seja, h pouca disperso.

### Varincia alta:
Indica que os dados esto mais espalhados em relao  mdia, ou seja, h maior disperso.

O desvio padro  muitas vezes mais fcil de interpretar do que a varincia porque ele est na mesma unidade de medida dos dados originais. Por exemplo, se seus dados esto em metros, o desvio padro tambm estar em metros, enquanto a varincia estaria em metros quadrados.

````python
# Variana em python

average # 424.16

diffs = 0

for i in numbers:
  diffs += (i - average) ** 2

variance = diffs / (len(numbers) -1 )
variance # 105293.24686868684
print(f'The variance of numbers: {variance:.2f}')
````
