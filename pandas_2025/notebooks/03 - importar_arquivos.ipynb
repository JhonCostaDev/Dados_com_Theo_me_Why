## Dataframe a partir de um link html

```python
# Url da wikipedia sobre o estado do Ceará
url = 'https://en.wikipedia.org/wiki/Cear%C3%A1'

# O read_html retorna uma lista com todas as tabelas que ele consegue recuperar
# do link
dfs = pd.read_html(url)

# Verificando o tamanho da lista
len(dfs)
```

```python
# Registro populacional
historico_populacional = dfs[3]

# Ranking das maiores cidades
maiores_cidades = dfs[4]
```
