# Variança

Na área da estatística, a variância é uma medida de dispersão que indica o quão espalhados estão os valores de um conjunto de dados em relação à sua média (ou valor esperado). Em outras palavras, ela quantifica a variabilidade dos dados.

## Conceito Intuitivo:

Imagine que você tem dois conjuntos de notas de alunos. Ambos os conjuntos podem ter a mesma média, mas em um conjunto as notas podem estar muito próximas da média (por exemplo, a maioria dos alunos tirou notas entre 7 e 8), enquanto no outro conjunto as notas podem estar mais dispersas (alguns alunos tiraram notas muito baixas e outros muito altas). A variância ajuda a distinguir essas situações, indicando que o segundo conjunto tem uma variância maior.

## Como a Variância é Calculada:

O cálculo da variância envolve os seguintes passos:

1. Calcular a média (μ) do conjunto de dados.
2. Para cada valor (x) no conjunto de dados, calcular a diferença entre esse valor e a média (x - μ). Este valor é chamado de desvio.
3. Elevar ao quadrado cada um desses desvios ((x - μ)²). Isso é feito para garantir que as diferenças negativas e positivas contribuam para a medida de dispersão e para dar mais peso aos desvios maiores.
4. Calcular a média desses desvios quadrados.

<hr>

## Fórmulas:

Existem duas fórmulas principais para calcular a variância, dependendo se você está trabalhando com uma população inteira ou com uma amostra:

### Variância Populacional (σ²):

```bash
σ² = Σ (xᵢ - μ)² / N
```
Onde:

* σ² (sigma ao quadrado) representa a variância populacional.
* Σ (sigma maiúsculo) indica a soma.
* xᵢ representa cada valor individual na população.
* μ (mi) representa a média da população.
* N representa o número total de elementos na população.


<hr>

### Variância Amostral (s²):

```bash
s² = Σ (xᵢ - x̄)² / (n - 1)

```
Onde:

* s² representa a variância amostral.
* Σ indica a soma.
* xᵢ representa cada valor individual na amostra.
* x̄ (x-barra) representa a média da amostra.
* n representa o número total de elementos na amostra.
O denominador é (n - 1) em vez de n para fornecer uma estimativa não viesada da variância populacional a partir da amostra (correção de Bessel).


## Interpretação da Variância:

### Variância baixa:
Indica que os dados estão agrupados próximos à média, ou seja, há pouca dispersão.

### Variância alta:
Indica que os dados estão mais espalhados em relação à média, ou seja, há maior dispersão.
Relação com o Desvio Padrão:

A variância está intimamente relacionada com o desvio padrão. O desvio padrão (σ para população, s para amostra) é simplesmente a raiz
quadrada da variância:

* Desvio Padrão Populacional (σ) = √σ²

* Desvio Padrão Amostral (s) = √s²

O desvio padrão é muitas vezes mais fácil de interpretar do que a variância porque ele está na mesma unidade de medida dos dados originais. Por exemplo, se seus dados estão em metros, o desvio padrão também estará em metros, enquanto a variância estaria em metros quadrados.

````python
# Variança em python

average # 424.16

diffs = 0

for i in numbers:
  diffs += (i - average) ** 2

variance = diffs / (len(numbers) -1 )
variance # 105293.24686868684
print(f'The variance of numbers: {variance:.2f}')
````

O calculo da variança utilizando apenas python pode ser complicado e um pouco confunso se você não tiver uma boa base de estatística. Os métodos de lista não ajudam muito nesse quesito. Por isso utilizaremos Pandas para facilitar nosso trabalho.

````python
import pandas as pd

# transformando a lista em uma Series

series_numbers = pd.Series(numbers)
series_numbers
````

````python
# Métodos de séries

series_numbers.mean() # media
series_numbers.var()  # Variança = 105293.24686868684
````
