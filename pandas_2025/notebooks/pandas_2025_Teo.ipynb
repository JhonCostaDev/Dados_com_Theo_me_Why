```python
import pandas as pd
import random

# idades = []
# for i in range(0, 100):
#   idades.append(random.randint(1, 80))
# Gerar lista de idades
idades = [random.randint(1, 80) for x in range(100)]

```

```python
# COnverti a lista para uma serie pandas
serie_idade = pd.Series(idades)
serie_idade
```

```python
# o index de uma serie funciona como a key de um dicionario
# portanto quando modificamos a posição dos elementos de uma
# serie, como por exemplo ordenando-a, o indice muda de posição
# junto do elemento
```

```python
# ordenar do menor para o maior

serie_idade = serie_idade.sort_values()
```

## Acessando independente do indice

```python
serie_idade.iloc[2] # terceiro elemento
serie_idade.iloc[:3] # Tres primeiros elementos
serie_idade.iloc[-1] # ultimo elemento
```

```python
# duas listas com a mesma quantidade de elementos

idades_funcionario = [random.randint(18, 40) for  x in range(15)]
nomes_funcionario = [
    "Ana", "Bruno", "Carlos", "Daniela", "Eduardo",
    "Fernanda", "Gustavo", "Helena", "Igor", "Juliana",
    "Karla", "Leonardo", "Marina", "Nicolas", "Olívia"
]
```

```python
# transformando as duas listas em um dicionario

funcionario = dict(zip(nomes_funcionario, idades_funcionario))
funcionario
```

```python
# Usandas pandas para criar uma serie com duas listas


funcionario_serie = pd.Series(idades_funcionario, index=nomes_funcionario)
funcionario_serie
```

```python
# Buscando numa serie
funcionario_serie[['Nicolas']]
```

```python
# Buscando por posicao

funcionario_serie.iloc[[-1]]
```

```python
funcionario_serie.get('Daniela')
```

## DataFrames

```python
serie_idades = pd.Series(idades_funcionario)
serie_nomes = pd.Series(nomes_funcionario)
```

```python
# Dataframe vazio

df = pd.DataFrame()
```

```python
df['idades'] = serie_idades
df['nomes'] = serie_nomes
df
```
